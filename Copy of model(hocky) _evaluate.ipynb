```python
import os
import cv2
import tensorflow
import keras
import time
import tensorboard
import numpy as np
from random import shuffle
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Reshape
from tensorflow import keras
from keras.models import load_model
from tensorflow.keras import Input
from keras.callbacks import TensorBoard
from tensorflow.keras.callbacks import TensorBoard
```

load hockyFight dataset numpy array

```python
#data = np.load('/content/dataset_hocky.npy', allow_pickle=True)
data = np.load('/content/drive/MyDrive/CSE299(CNN BiLSTM)/numpy/dataset_hocky.npy', allow_pickle=True)
```

```python
train, test = train_test_split(data, train_size= 0.9, shuffle=True)
```

```python
train[0][1]
```

```python
X = np.array([i[0] for i in train]).reshape(-1, 10, 100, 100, 3)
Y = np.array([i[1] for i in train])

x_valid = np.array([i[0] for i in test]).reshape(-1, 10, 100, 100, 3)
y_valid = np.array([i[1] for i in test])
```

```python
X = X.astype('float32')/255
x_valid = x_valid.astype('float32')/255
```

```python
print(Y.shape)
print(y_valid.shape)
print(X.shape)
print(x_valid.shape)
print(y_valid[7])
```

```python
x_try = X[:20]
y_try = Y[:20]
print(len(x_try), len(y_try))
```

model load

```python
cnn = load_model('/content/drive/MyDrive/CSE299(CNN BiLSTM)/Model(h5 file)/CNN-BiLSTM.h5')
cnn.summary()
```

```python
#tb.callback=tensorflow.keras.callbacks.TensorBoard(log_dir="logs/",histogram_freq=1)

opt=keras.optimizers.SGD(learning_rate=0.01)
cnn.compile(loss='sparse categorical crossentropy', optimizer=opt, metrics=['accuracy'])
NAME= 'CNN-BiLSTM-{}'.format(int(time.time()))

tensorboard = TensorBoard(log_dir='/content/drive/MyDrive/CSE299(CNN BiLSTM)/log/{}'.format(NAME))
```

train

```python
cnn.fit(X,Y,epochs=10,validation_data=(x_valid,y_valid),batch_size=16, verbose=1)
```

```python
cnn.save('CNN-BiLSTM.h5', overwrite=True, include_optimizer=True)
```
